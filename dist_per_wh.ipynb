```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("/home/zeal/Documents/6th sem/inv. mangmnt sy/large-scale route optimisation/top_material_ids.csv")
```

```python
# Group by Warehouse_x, Warehouse_ID_x, and Category
grouped = df.groupby(['Destination', 'Category']).size().reset_index(name='Quantity')
```

```python
# Display the grouped DataFrame
print(grouped)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualization
sns.set(style="whitegrid")

# Create a pivot table for the heatmap
pivot_table = grouped.pivot(index='Destination', columns='Category', values='Quantity')

# Create a heatmap
plt.figure(figsize=(14, 8))
heatmap = sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="YlGnBu")

# Set plot titles and labels
plt.title('Heatmap of Category Distribution Across Different Warehouses')
plt.xlabel('Category')
plt.ylabel('Warehouse')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualization
sns.set(style="whitegrid")

# Create a FacetGrid
g = sns.FacetGrid(grouped, col="Destination", col_wrap=3, height=4, aspect=1.5)
g.map(sns.barplot, "Category", "Quantity")

# Set plot titles and labels
g.set_axis_labels("Category", "Quantity")
g.set_titles("{col_name} Warehouse")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Distribution of Categories Across Different Warehouses')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualization
sns.set(style="whitegrid")

# Create a stacked bar plot
plt.figure(figsize=(14, 8))
stacked_bar_plot = grouped.pivot(index='Destination', columns='Category', values='Quantity').plot(kind='bar', stacked=True, figsize=(14, 8))

# Set plot titles and labels
plt.title('Distribution of Categories Across Different Warehouses')
plt.xlabel('Warehouse')
plt.ylabel('Quantity')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualization
sns.set(style="whitegrid")

# Get unique warehouses
unique_warehouses = grouped['Destination'].unique()

# Create bar plots for each warehouse
for warehouse in unique_warehouses:
    warehouse_data = grouped[grouped['Destination'] == warehouse]

    plt.figure(figsize=(14, 8))
    bar_plot = sns.barplot(data=warehouse_data, x='Category', y='Quantity', dodge=True)

    # Rotate x-axis labels for better readability
    bar_plot.set_xticklabels(bar_plot.get_xticklabels(), rotation=45, horizontalalignment='right')

    # Set plot titles and labels
    plt.title(f'Distribution of Categories in {warehouse}')
    plt.xlabel('Category')
    plt.ylabel('Quantity')
    plt.legend(title='Warehouse ID')
    plt.tight_layout()

    # Show the plot
    plt.show()
```
